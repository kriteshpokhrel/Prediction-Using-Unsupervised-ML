```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import *
```

```python
iris_data=pd.read_csv('C:/Users/krite/Desktop/Iris.csv')
iris_data = iris_data.iloc[:,1:]
iris_data.head()
```

# Data is to be classified into different species based on first 4 columns and the species label will not be used for training our data because of which it is called Unsupervised Machine Learning.

Finding the optimum number of clusters for k-means classification

```python
from sklearn.cluster import KMeans
X=iris_data.iloc[:,:-1].values
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init = 'k-means++', 
                    max_iter = 300, n_init = 10,random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.title('Estimating Optimum Number of clusters using Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()


#From the above plot, we can see that the optimum number of clusters is 3. When the numbers of clusters are increased beyond this there is no significant decrease in the Within Cluster Sum of Squares (WCSS)
```

```python
#Choosing number of clusters as 3
kmeans=KMeans(n_clusters=3,random_state=0)
y_kmeans=kmeans.fit_predict(X)
```

```python
y_kmeans
```

# Visualizing the clusters on basis of every pair of features

```python
# Visualising the clusters - On the first two columns
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
